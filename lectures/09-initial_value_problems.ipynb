# Initial Value Problems

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Forward (explicit) Euler's $1^{st}$ order method:

```python
def forward_euler(f, x_0, x_n, y_0, n): # Runge-Kutta 1st order method
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    for i in range(n-1):
        y[i+1] = y[i] + h * f(x[i], y[i])
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = forward_euler(f, 0.0, 3200.0, 1.0, 10)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="Forward Euler's method")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

### Runge-Kutta $2^{nd}$ order method:

```python
def rk_2(f, x_0, x_n, y_0, n):
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    for i in range(n-1):
        k_1 = f(x[i], y[i])
        #k_2 = f(x[i] + h / 2.0, y[i] + h / 2.0 * k_1) 
        #y[i+1] = y[i] + h * k_2 # rectangular rule
        k_2 = f(x[i] + h, y[i] + h * k_1) 
        y[i+1] = y[i] + h / 2.0 * (k_1 + k_2) # trapezoid rule (Heun's predictor-corrector method) 
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = rk_2(f, 0.0, 3200.0, 1.0, 10)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="RK2 method")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

### Runge-Kutta $3^{rd}$ order method:

```python
def rk_3(f, x_0, x_n, y_0, n):
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    for i in range(n-1):
        k_1 = f(x[i], y[i])
        k_2 = f(x[i] + h / 2.0, y[i] + h / 2.0 * k_1)
        k_3 = f(x[i] + h, y[i] + h * (-k_1 + 2.0 * k_2))
        y[i+1] = y[i] + h * (k_1 / 6.0 + 2.0 * k_2 / 3.0 + k_3 / 6.0) # Simpson's 1/3 rule
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = rk_3(f, 0.0, 3200.0, 1.0, 10)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="RK3 method")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

### Runge-Kutta $4^{th}$ order method:

```python
def rk_4(f, x_0, x_n, y_0, n):
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    for i in range(n-1):
        k_1 = f(x[i], y[i])
        k_2 = f(x[i] + h / 2.0, y[i] + h / 2.0 * k_1)
        k_3 = f(x[i] + h / 2.0, y[i] + h / 2.0 * k_2)
        k_4 = f(x[i] + h, y[i] + h * k_3)
        y[i+1] = y[i] + h / 6.0 * (k_1 + 2.0 * k_2 + 2.0 * k_3 + k_4)
        #k_2 = f(x[i] + h / 3.0, y[i] + h / 3.0 * k_1)
        #k_3 = f(x[i] + 2.0 * h / 3.0, y[i] + h * (-k_1 / 3.0 + k_2))
        #k_4 = f(x[i] + h, y[i] + h * (k_1 - k_2 + k_3))
        #y[i+1] = y[i] + h * (k_1 / 8.0 + 3.0 * k_2 / 8.0 + 3.0 * k_3 / 8.0 + k_4 / 8.0) # Simpson's 3/8 rule
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = rk_4(f, 0.0, 3200.0, 1.0, 10)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="RK4 method")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

### Backward (implicit) Euler's $1^{st}$ order method:

```python
def newton_raphson(f, x0, h=1.0e-8, tol=1.0e-8, n_max=100):
    n = 0
    while n <= n_max:
        x1 = x0 - f(x0) / ((f(x0 + h) - f(x0)) / h)
        if f(x1) == 0 or abs(x1 - x0) < tol:
            return x1
        x0 = x1
        n += 1
    raise RuntimeError('no root found using', n_max, 'iterations')
```

```python
def backward_euler(f, x_0, x_n, y_0, n): # implicit method
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    for i in range(n-1):
        y[i+1] = newton_raphson(lambda z: z - y[i] - h * f(x[i+1], z), y[i])
    return x, y
```

```python
def f(x, y):
    return -100.0 * y + 100.0

y0 = 5.0

x_1, y_1 = forward_euler(f, 0.0, 0.5, y0, 25)
x_2, y_2 = backward_euler(f, 0.0, 0.5, y0, 25)

x_exact = np.linspace(0.0, 0.5, 1000)
y_exact = (y0 - 1.0) * np.exp(-100.0 * x_exact) + 1.0
    
fig, ax = plt.subplots()
ax.plot(x_1, y_1, '-', color='red', label="Forward Euler's method")
ax.plot(x_2, y_2, 'o', color='blue', label="Backward Euler's method")
ax.plot(x_exact, y_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

### Multistep methods:

```python
def leap_frog(f, x_0, x_n, y_0, n): 
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    y[1] = y[0] + h * f(x[0], y[0]) 
    for i in range(1, n-1):
        y[i+1] = y[i-1] + 2.0 * h * f(x[i], y[i])
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = leap_frog(f, 0.0, 3200.0, 1.0, 10)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="Leap-frog method")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

```python
def adams_bashforth(f, x_0, x_n, y_0, n): 
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    y[1] = y[0] + h * f(x[0], y[0]) 
    for i in range(1, n-1):
        y[i+1] = y[i] + 3.0 / 2.0 * h * f(x[i], y[i]) - 1.0 / 2.0 * h * f(x[i-1], y[i-1])
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = adams_bashforth(f, 0.0, 3200.0, 1.0, 10)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="Adams-Bashforth method")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

```python
def adams_moulton(f, x_0, x_n, y_0, n): 
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    y[1] = y[0] + h * f(x[0], y[0]) 
    for i in range(1, n-1):
        y[i+1] = newton_raphson(lambda z: z - y[i] - h * (5.0 / 12.0 * f(x[i+1], z) + 2.0 / 3.0 * \
            f(x[i], y[i]) - 1.0 / 12.0 * f(x[i-1], y[i-1])), y[i])
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = adams_moulton(f, 0.0, 3200.0, 1.0, 10)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="Adams-Moulton method")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

```python
def predictor_corrector(f, x_0, x_n, y_0, n):
    x, h = np.linspace(x_0, x_n, n, retstep=True)
    y = np.zeros(n)
    y[0] = y_0
    y[1] = y[0] + h * f(x[0], y[0]) 
    for i in range(1, n-1):
        q = y[i] + 3.0 / 2.0 * h * f(x[i], y[i]) - 1.0 / 2.0 * h * f(x[i-1], y[i-1])
        y[i+1] = y[i] + h * (5.0 / 12.0 * f(x[i+1], q) + 2.0 / 3.0 * f(x[i], y[i]) - 1.0 / 12.0 * f(x[i-1], y[i-1]))
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = predictor_corrector(f, 0.0, 3200.0, 1.0, 10)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="Predictor-corrector method")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```

### Bulirsch-Stoer algorithm:

```python
def bulirsch_stoer(f, x_0, x_n, y_0, n, m=5): 
    x = np.linspace(x_0, x_n, n)
    y = np.zeros(n)
    y[0] = y_0
    for i in range(n-1):
        T = np.zeros((m, m))
        T[0, 0] = leap_frog(f, x[i], x[i+1], y[i], 2)[1][-1]
        for k in range(1, m):
            T[k, 0] = leap_frog(f, x[i], x[i+1], y[i], 2 * (k+1))[1][-1]
            for j in range(1, k+1):
                T[k, j] = T[k, j-1] + 1.0 / (4**j - 1) * (T[k, j-1] - T[k-1, j-1]) 
        y[i+1] = T[m-1, m-1]
    return x, y
```

```python
def f(t, N):
    return -np.log(2.0) / 1600.0 * N

t, N = bulirsch_stoer(f, 0.0, 3200.0, 1.0, 5, 4)

t_exact = np.linspace(0.0, 3200.0, 1000)
N_exact = 1.0 * np.exp(-np.log(2.0) / 1600.0 * t_exact)
    
fig, ax = plt.subplots()
ax.plot(t, N, 'o', color='red', label="Burlisch-Stoer")
ax.plot(t_exact, N_exact, '--', color='black', label="Analytical solution")
ax.set_xlabel("t (years)")
ax.set_ylabel("N (t)")
ax.legend()
```
